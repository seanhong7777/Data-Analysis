# list를 활용한 Vector 연산

````
def add_vector(v1, v2):
    """
    원소의 개수가 같은 list 2개를 argument로 전달받아서, 
    같은 인덱승의 원소들끼리(elment-wise) 더한 값들로 이루어진 list를 생성해서 리턴하는 함수
    :param v1: 숫자들을 원소로 갖는 1차원 list
    :param v2: 숫자들을 원소로 갖는 1차원 list
    """
    if len(v1) != len(v2):  # v1과 v2의 원소의 개수가 다르면
        raise ValueError('v1과 v2는 기리가 같아야 함')

    return [x + y for x, y in zip(v1, v2)]
    # return [v1[i] + v2[i] for i in range(len(v1))]
````

````
v1 = [1, 2]
v2 = [2, 1]
add_vector(v1, v2)
````

````
v1 = [1, 2, 3]
v2 = [2, 3, 4]
add_vector(v1, v2)
````

````
def scalar_multiply(v: list, n: float) -> list:
    """
    숫자들의 리스트 v와 숫자 n을 argument로 전달받아서, 
    리스트의 각 원소들에 n을 곱한 값들로 이루어진 리스트를 반환

    : param v: 숫자들의 list
    : param n: 숫자
    : return : list    
    """
    return [x * n for x in v]
````

````
vec = [1, 2]
scalar_multiply(vec, 2) #> [2, 4]
````

````
def vector_multiply(v1: list, v2: list) -> list:
    """v1 = [x1, x2, x3, ..., xn], v2 = [y1, y2, y3, ..., yn]일 때, 
    [x1 * y1, x2 * y2, ..., xn * yn]을 리턴하는 함수

    :param v1: 숫자들의 list
    :param v2: 숫자들의 list
    :return : list
    """

    if len(v1) != len(v2):
        raise ValueError('v1과 v2는 같은 길이여야 함')

    return [x * y for x, y in zip(v1, v2)]
````

````
# vector_multiply 테스트
v1, v2 = [1, 2], [2, 3]
vector_multiply(v1, v2)
````

````
def vector_dot(v1: list, v2: list) -> float:
    """
    v1 = [x1, x2, x3, ..., xn], v2 = [y1, y2, y3, ..., yn]일 때,
    x1 * y1 + y2 + y2, x3 + y3 + ... + xn * yn을 리턴하는 함수

    : param v1 : 숫자들의 list
    : param v2 : 숫자들의 list
    : return : number
    """

    if len(v1) != len(v2):
        raise ValueError('v1과 v2는 같은 길이여야 함')
    
    # return sum([x * y for x, y in zip(v1, v2)])
    return sum(vector_multiply(v1, v2))
````

````
# vector_dot 테스트
v1, v2 = [1, 2], [2, 3]
vector_dot(v1, v2)
````

````
def vector_norm(v: list) -> float:
    """vector = [x1, x2, x3, ..., xn] 일 때,
    sqrt(x1**2, x2**2, x3**3+ ... + xn***2)을 리턴하는 함수 
    
    :param v: 숫자들을 원소로 갖는 list
    :return: 숫자   
    """
    
    # return math.sqrt(sum([x**2 for x in v]))
    return math.sqrt(vector_dot(v, v))
````

# NumPy(Numerical Python)

* 배열(array: 숫자들의 리스트)을 쉽고, 빠르게 연산(+, -, *, /, ..)하기 위한 라이브러리
* NumPy 패키지의 대부분의 기능(함수, 메서드)들은 반복문의 기능을 가지고 있음 
* NumPy 패키지의 대부분의 기능(함수, 메서드)들은 같은 인덱스의 원소들끼리(element-wise) 연산을 수행하는 경우가 많음
* np.ndarray 클래스: NumPy 패키지의 기본 데이터 타입. n-dimensional array(n차원 배열)

````
import numpy as np
# numpy 패키지를 np라는 별명으로 import
````

````
print(np.__version__)   # numpy 패키지의 버전 
````

````
l = [1, 2, 3]  # list 클래스
print(l, type(l))
````

````
arr1 = np.array([1, 2, 3])  # np.ndarray 클래스
print(arr1, type(arr1))
````

````
arr2 = np.array([3, 4, 5])  
print(arr2, type(arr2))
````

````
print(arr1 + arr2)  # 같은 인덱스의 원소들끼리 + 연산을 수행
````

````
print(arr1 - arr2)
````

````
print(arr1 * arr2)
````

````
print(arr1 / arr2)
````

````
print(arr1 * 3)
````

````
print(3 * arr1)
````

````
print(arr1)
print(arr2)
print(np.dot(arr1, arr2))  # np.dot(a, b) 함수
print(arr1.dot(arr2)) # a.dot(b) 함수
print(arr1 @ arr2)  # @: Python f3.5 버전 이상에서 사용 가능한 연산자
````

np.ndarray의 속성들(attribute): 차원, 모양, 크기, 데이터 타입

````
print(arr1)
print('dimension(차원):', arr1.ndim)
print('shape(모양):', arr1.shape)
print('size(크기):', arr1.size)
print('data type:', arr1.dtype)
````

````
print(np.dot(arr1, arr2))
# X @ Y 연산: x의 column의 개수와 y의 row의 개수가 같은 경우에 연산 가능
# (n, m) @ (m, k)
# X @ Y != Y @ X  교환 법칙이 성립하지 않음
````

````
arr1 = np.array([[1, 2, 3],
                 [4,5, 6]])  # shape: (2, 3)
arr2 = np.array([[1, 2],
                 [3, 4],
                 [5, 6]])      # shape: (3, 2)
print(np.dot(arr1, arr2))
# (2, 3) @ (3, 2) = (2, 2)
````
