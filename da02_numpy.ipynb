# np.ndarray 객체 생성 함수들

**``np.array()`` 함수**

```
import numpy as np
```

**``np.array(array-like object)`` 함수
* array-like-object: list, tuple, array

```
arr = np.array([1, 2, 3])
print(arr)
```

```
arr = np.array([2 * x for x in range(5)])
print(arr)
```

**`np.arrange(start, end, step)` 함수**

* array using range
* 1-d array를 생성
* range(end), range(start, end), range(start, end, step)

```
print(np.arange(5))
print(np.arange(1, 6))
print(np.arange(0, 10, 2))
```

**`ndarray.reshape(shape)` 메서드**

* `ndarray.reshape(tuple)`
* `ndarray.reshape(x, y, ...)`

```
arr_1d = np.arange(6)
print(arr_1d)

arr_2d = arr_1d.reshape((2, 3))  # arr_1d.reshape(2, 3)
print(arr_2d)
```

```
print(np.arange(6).reshape(3, 2))
```

**`np.zeros(shape)`함수**

* 모든 원소가 0인 ndarray를 생성

```
arr = np.zeros((2, 3))
print(arr)
```

```
arr.dtype
```

```
print(np.zeros((2, 3), dtype=int))
```

```
print(np.zeros(2, 3))
# TypeError가 발생
```

**`np.ones(shape)` 함수**

* 모든 원소가 1로 이루어진 ndarray를 생성


```
arr = np.ones((2, 3))
print(arr)
```

**`np.eye()` 함수**

* 대각선에는 1, 그 이외에는 0으로 채워진 2차원 2-d ndarray를 생성

```
arr = np.eye(3, dtype=int)
print(arr)  #> (3, 3)
```

```
print(np.eye(3, 4, dtype=int))
```

```
print(np.eye(4, k=1, dtype=int))  #> (4, 4)
```

**`np.random.rand(d0, d1, ... dn)` 함수**:

0 <= r < 1 난수들로 이루어진 ndarray를 생성

```
arr  = np.random.rand(5)
print(arr)
```

```
arr  = np.random.rand(3, 2)
print(arr)
```

**`np.random.randn(d0, d1, ...., dn)` 함수**: 

평균이 0이고, 표준편차가 1인 표준정규분포(standard normal distribution)를 따르는 난수들의 배열을 생성

```
arr = np.random.randn(10)
print(arr)
```

```
arr = np.random.randn(4, 3)
print(arr)
```

**`np.random.randint(low, high, size, dtype)` 함수** 

* low <= r < high 범위의 정수 난수들로 이루어진 배열을 생성
* size 파라미터에는 ndarray의 shape을 argument로 전달해야 

```
arr = np.random.randint(1, 4, size=10)
print(arr)
```

```
arr = np.random.randint(1, 4, size=(3, 4))
print(arr)
```

```
import matplotlib.pyplot as plt
```

```
x = np.arange(-3, 3, 1)
print(x)

y = x ** 2
print(y)
```

```
plt.plot(x, y)
plt.grid()
plt.show()
```

```
x = np.arange(-3, 3, 0.01)
print(x[:10])
print(x[-10:])

y = x ** 2
```

```
plt.plot(x, y)
plt.grid()
plt.show()
```

```
x = np.arange(-1.1, 1.1, 0.01)
y = x ** 3 - x

plt.plot(x, y)
plt.grid()
plt.show()
```

```
x = np.arange(-3, 3, 0.001)
y = 2 ** x

plt.plot(x, y)
plt.grid()
plt.show()
```

```
x = np.arange(-3, 3, 0.001)
y = 0.5 ** x

plt.plot(x, y)
plt.grid()
plt.show()
```

```
x = np.arange(-3, 3, 0.001)
y = np.exp(x)
y2 = 2 ** x

plt.plot(x, y)
plt.plot(x, y2)
plt.grid()
plt.show()
```

```
x = np.arange(0.001, 3, 0.001)
y = np.log(x)

plt.plot(x, y)
plt.grid()
plt.show()
```

```
# 지수 함수 vs 로그 함수
x1 = np.arange(-20, 3, 0.001)
y1 = np.exp(x1)

x2 = np.arange(0.00001, 10, 0.001)
y2 = np.log(x2)

x3 = np.arange(-10, 10, 0.001)
y3 = x1  # y = x

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x1, y3)

plt.grid()
plt.axis('square')
plt.show()
```
