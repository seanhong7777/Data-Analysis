```
# 연도별 lifeExp의 평균 출력 
lifeExp_by_year = []  # 연도별 기대수명을 저장할 리스트

years = gapminder['year'].unique()  # 중복되지 않는 year 값들
for y in years:   # (1952, 1957, 1962, ..., 2002, 2007)
    subset = gapminder[gapminder['year'] == y]  # 각 연도 부분집합
    avg = subset['lifeExp'].mean()
    lifeExp_by_year.append(avg)

lifeExp_by_year
```

```
s = pd.Series(data=lifeExp_by_year, index=years)
s
```

```
# 연도별 전세계 인구 출력 
pop_by_year = []  # 연도별 전세계 인구를 저장할 리스트
for y in years:
    subset = gapminder[gapminder['year'] == y]
    pop = subset['pop'].sum()
    pop_by_year.append(pop)
                       
s = pd.Series(data=pop_by_year, index=years)
s
```

```
# 연도별 전세계 인구 시각화
plt.plot(s.index, s.values, 'o-')
plt.grid()
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```
# 연도별 대륙별 전세계 인구
continents = gapminder['continent'].unique()
years = gapminder['year'].unique()

for c in continents:
    # print(f'**** {c} ****')
    gapminder_by_continent = gapminder[gapminder['continent'] == c] # 대륙별 부분집합
    total_pop = []  # 해당 대륙의 연도별 대륙별 인구를 저장할 리스트 
    for y in years:
        # 해당 대륙에서 해당 연도의 부분 집합 
        subset = gapminder_by_continent[gapminder_by_continent['year'] == y]
        sum = subset['pop'].sum()
        total_pop.append(sum)
    s = pd.Series(data=total_pop, index=years)
    # print(s, '\n')
    plt.plot(s.index, s.values, 'o-', label=c)

plt.legend()
plt.show()
```
