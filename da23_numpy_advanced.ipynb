* ndarray 이어붙이기(concatenate) 
    * `np.concatenate([a1, a2, ...], axis)` : row concatenation (행 이어붙이기)
    * `np.r_[a1, a2]` : column concatenation (열 이어붙이기)

```
a1 = np.arange(6).reshape((2, 3))
a1
```

```
a2 = np.arange(6, 12).reshape((2, 3))
a2
```

```
np.concatenate([a1, a2])  # axis=0 (기본값)
```

```
np.concatenate([a1, a2], axis=1) 
```

```
np.r_[a1, a2]
```

```
np.c_[a1, a2]
```

```
X = np.array([[1, 1],
             [2, 2],
             [3, 3]])
X
```

```
W = np.array([[2],
              [2],
              [2]])
W
```

```
# X @ W 에러 발생
X_bias = np.c_[np.ones((3, 1), dtype=np.int_), X]
X_bias
```

내적을 하려면 (3, 2) @ (3, 1) 은 불가능 

내적(행렬들의 닷 프로덕트)을 할 수 있도록 X를 변형을 시켜줌


```
X_bias @ W
```

# Broadcast 

* ndarray의 산술 연산은 같은 위치(인덱스)의 원소들끼리(element-wise) 계산하는 것이 기본임
* 다른 모양(shape)의 배열들 사이의 산술 연산(+, -, *, /, //, %, ...)을 수행하는 방법 -> Broadcasting

```
a = np.arange(5)
b = np.arange(4)
a + b
#> ValueError: operands could not be broadcast together with shapes (5,) (4,) 
```

```
a = np.arange(5)
a
```

```
a + 1
```

```
a = np.arange(6).reshape((2, 3))
a
```

```
b = np.array([1, 2, 3])
b
```

```
a + b
```

```
a = np.ones((2, 3), dtype=np.int_)
a
```

```
b = np.array([[1], [2]])
b
```

```
a + b
```

```
# min-max normalization(최소-최대 정규화): (X - min(X)) / (max(X) - min(X))
# 최솟값 0, 최댓값 1, 나머지 0과 1 사이의 숫자로 변환 
np.random.seed(1)
x = np.random.randint(1, 10, 10) 
x 
```

```
# 1차원 배열 x를 min-max normalization 적용
n = (x - np.min(x)) / (np.max(x) - np.min(x))
n
```
